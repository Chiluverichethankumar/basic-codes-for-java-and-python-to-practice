```python
from bs4 import BeautifulSoup
import requests
```

```python
url1="https://www.scrapethissite.com/pages/forms"
page=requests.get(url1)
print(page)
```

```python
soup1=BeautifulSoup(page.text,'html')
print(soup1)
```

```python
soup1.find('div')
```

```python
soup1.find_all('div')
```

```python
soup1.find_all('div', class_="col-md-12")
```

```python
soup1.find_all('p')
```

```python
soup1.find_all(class_="lead")
```

```python
soup1.find_all(class_="lead").text
```

```python
soup1.find(class_="lead").text
```

```python
soup1.find(class_="lead").text.strip()
```

```python
soup1.find("th")
```

```python
soup1.find_all("th")
```

```python
soup1.find_all("td")
```

```python
soup.find("th").text.strip()
```
